```python
from pyspark.sql.functions import avg, count, date_format

# STEP 1: Read Silver table
silver_df = spark.read.format("delta") \
    .load("/Volumes/workspace/default/silver/weather_clean")

# STEP 2: Aggregate business metrics
gold_df = silver_df.groupBy(
    "city",
    date_format("event_time", "yyyy-MM-dd").alias("date")
).agg(
    avg("temperature").alias("avg_temperature"),
    avg("humidity").alias("avg_humidity"),
    count("*").alias("record_count")
)

# STEP 3: Write Gold Delta table
gold_df.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/workspace/default/gold/weather_analytics")
```
