```python
bronze_df = spark.read \
    .option("recursiveFileLookup", "true") \
    .json("/Volumes/workspace/default/raw-weather")

bronze_df.printSchema()
bronze_df.show(5)


bronze_df.write.format("delta") \
    .mode("append") \
    .save("/Volumes/workspace/default/bronze/weather")
```
